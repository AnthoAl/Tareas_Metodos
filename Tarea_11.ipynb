1. Encuentre las primeras dos iteraciones del m√©todo de Jacobi para los siguientes sistemas lineales, por

$$
\begin{aligned}
    3x_1 - x_2 + x_3 &= 1, \\
    3x_1 + 6x_2 + 2x_3 &= 0, \\
    3x_1 + 3x_2 + 7x_3 &= 4. \\
\end{aligned}

```python
import numpy as np


def resolver_jacobi(A, b, x0, tol, max_iter):
    
    A = A.astype(float)
    b = b.astype(float)
    x = x0.astype(float)
    n = len(b)

    print(f"Iteraci√≥n 0: {x}")

    for k in range(1, max_iter + 1):
        x_nuevo = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x_nuevo[i] = (b[i] - s) / A[i][i]

        error = np.linalg.norm(x_nuevo - x, ord=np.inf)
        x = x_nuevo

        print(f"Iteraci√≥n {k}: {x.round(4)}")

        if error < tol:
            print(f"\nConvergencia alcanzada en {k} iteraciones.")
            return x

    return x
```

```python
A_a = np.array([[3, -1, 1], [3, 6, 2], [3, 3, 7]])

b_a = np.array([1, 0, 4])

x_inicial = np.zeros(len(b_a))

print("Resolviendo el sistema a:")
resultado = resolver_jacobi(A_a, b_a, x_inicial, tol=1e-6, max_iter=2)
```

```python
import numpy as np

def resolver_gauss_seidel(A, b, x0, tol, max_iter):
    A = A.astype(float)
    b = b.astype(float)
    x = x0.astype(float)
    n = len(b)

    print(f"Iteraci√≥n 0: {x}")

    for k in range(1, max_iter + 1):
        x_anterior = np.copy(x)  
        
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x[i] = (b[i] - s) / A[i][i]

        error = np.linalg.norm(x - x_anterior, ord=np.inf)

        print(f"Iteraci√≥n {k}: {x.round(4)}")

        if error < tol:
            print(f"\nConvergencia alcanzada en {k} iteraciones.")
            return x

    return x
```
