2. Utilice la eliminación gaussiana con sustitución hacia atrás y aritmética de redondeo de dos dígitos para resolver
los siguientes sistemas lineales. No reordene las ecuaciones. (La solución exacta para cada sistema es 𝑥1 = −1,

a. $$-x_1 + 4x_2 + x_3 = 8$$

$$\frac{5}{3}x_1 + \frac{2}{3}x_2 + \frac{2}{3}x_3 = 1$$

$$2x_1 + x_2 + 4x_3 = 11$$

```python
import numpy as np

def redondeo_dos_dig(x):
    if x == 0:
        return 0.0
    return float(f"{x:.2g}")

def gaussian_elimination_2_digits(A, b):
    n = len(b)
    Ab = np.zeros((n, n + 1))
    for i in range(n):
        for j in range(n):
            Ab[i, j] = redondeo_dos_dig(A[i, j])
        Ab[i, n] = redondeo_dos_dig(b[i])

    print("Matriz inicial redondeada:\n", Ab)

    for i in range(n):
        pivot = Ab[i, i]
        if pivot == 0:
            raise ValueError("Cero en la diagonal.")
        
        for j in range(i + 1, n):
            factor = redondeo_dos_dig(Ab[j, i] / pivot)
            for k in range(i, n + 1):
                Ab[j, k] = redondeo_dos_dig(Ab[j, k] - redondeo_dos_dig(factor * Ab[i, k]))
        print("\n",Ab)

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        sum_ax = 0
        for j in range(i + 1, n):
            sum_ax = redondeo_dos_dig(sum_ax + redondeo_dos_dig(Ab[i, j] * x[j]))
        
        x[i] = redondeo_dos_dig(redondeo_dos_dig(Ab[i, n] - sum_ax) / Ab[i, i])
    
    return x

A_a = np.array([[-1, 4, 1], 
                [5/3, 2/3, 2/3], 
                [2, 1, 4]])
b_a = np.array([8, 1, 11])

print("--- RESOLVIENDO SISTEMA A ---")
sol_a = gaussian_elimination_2_digits(A_a, b_a)
print("\nSolución aproximada Sistema A:", sol_a)
```

3. Utilice el algoritmo de eliminación gaussiana para resolver, de ser posible, los siguientes sistemas lineales, y
determine si se necesitan intercambios de fila:

a. $$x_1 - x_2 + 3x_3 = 2$$

$$3x_1 - 3x_2 + x_3 = -1$$

$$x_1 + x_2 = 3$$

```python
def eliminacion_gaussiana_inter_filas(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)
    
    Ab = np.hstack([A, b.reshape(-1, 1)])
    
    print("Matriz aumentada inicial:\n", Ab)

    for i in range(n):
        max_row = np.argmax(np.abs(Ab[i:, i])) + i
        
        if Ab[max_row, i] == 0:
            raise ValueError("El sistema no tiene solución.")

        if max_row != i:
            Ab[[i, max_row]] = Ab[[max_row, i]]
            print(f"Intercambio de fila {i+1} con fila {max_row+1}")

        for j in range(i + 1, n):
            factor = Ab[j, i] / Ab[i, i]
            Ab[j, i:] = Ab[j, i:] - factor * Ab[i, i:]
            
        print(f"Estado tras paso {i+1}:\n", Ab)

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (Ab[i, n] - np.dot(Ab[i, i+1:n], x[i+1:n])) / Ab[i, i]
    
    return x

A_a = np.array([[1, -1, 3], 
                [3, -3, 1], 
                [1, 1, 0]])
b_a = np.array([2, -1, 3])

print("--- RESOLVIENDO SISTEMA A ---")
sol_a = eliminacion_gaussiana_inter_filas(A_a, b_a)
print("\nSolución Sistema A:", sol_a)
```

Para este sistema, es necesario realizar un intercambio de filas porque el segundo coeficiente de $x_2$ (segundo pivote) se hizo cero después de realizar el primer paso de eliminación, lo que impide continuar con el proceso y por ello, se intercambia la fila 2 con la fila 3.

$$-x_1 + 2x_3 = 3$$

En este caso, no fue necesario realizar intercambios de filas puesto que ninguno de los pivotes fue cero durante el proceso de eliminación gaussiana.

```python
def eliminacion_gaussiana_32bit(A, b):
    A = A.astype(np.float32)
    b = b.astype(np.float32)
    n = len(b)
    
    Ab = np.hstack([A, b.reshape(-1, 1)]).astype(np.float32)
    
    print("Matriz aumentada inicial:\n", Ab)

    for i in range(n):
        max_row = np.argmax(np.abs(Ab[i:, i])) + i
        
        if Ab[max_row, i] == 0:
            raise ValueError("El sistema no tiene solución.")

        if max_row != i:
            Ab[[i, max_row]] = Ab[[max_row, i]]
            print(f"Intercambio de fila {i+1} con fila {max_row+1}")

        for j in range(i + 1, n):
            factor = Ab[j, i] / Ab[i, i]
            Ab[j, i:] = Ab[j, i:] - factor * Ab[i, i:]
            
        print(f"Estado tras paso {i+1}:\n", Ab)

    x = np.zeros(n, dtype=np.float32)
    for i in range(n - 1, -1, -1):
        x[i] = (Ab[i, n] - np.dot(Ab[i, i+1:n], x[i+1:n])) / Ab[i, i]
    
    return x

A_a = np.array([[1/4, 1/5, 1/6], 
                [1/3, 1/4, 1/5], 
                [1/2, 1, 2]], dtype=np.float32)

b_a = np.array([9, 8, 8], dtype=np.float32)

print("--- RESOLVIENDO SISTEMA A ---")
sol_a = eliminacion_gaussiana_32bit(A_a, b_a)
print("\nSolución Sistema A:", sol_a)
```

5. Dado el sistema lineal:

$$ x_1 - x_2 + \alpha x_3 = -2 $$

$$ -x_1 + 2x_2 - \alpha x_3 = 3 $$

$$ \alpha x_1 + x_2 + x_3 = 2 $$

a. Encuentre el valor(es) de 𝛼 para los que el sistema no tiene soluciones.

b. Encuentre el valor(es) de 𝛼 para los que el sistema tiene un número infinito de soluciones.

c. Suponga que existe una única solución para una a determinada, encuentre la solución.

```python
import sympy as sp

def eliminacion_gaussiana_simbolica():
    alpha = sp.symbols('alpha')
    
    Ab = sp.Matrix([
        [1, -1, alpha, -2],
        [-1, 2, -alpha, 3],
        [alpha, 1, 1, 2]
    ])
    
    n = 3
    
    print("Matriz Aumentada Inicial:\n")
    sp.pprint(Ab)
    print("-" * 30)

    for i in range(n):
        if Ab[i, i] == 0:
            for j in range(i + 1, n):
                if Ab[j, i] != 0:
                    Ab.row_swap(i, j)
                    print(f"Intercambio de fila {i+1} con {j+1}")
                    break

        for j in range(i + 1, n):
            if Ab[i, i] != 0:
                factor = Ab[j, i] / Ab[i, i]
                Ab.row_op(j, lambda v, k: v - factor * Ab[i, k])
    
    Ab = sp.simplify(Ab)
    print("Matriz después de la eliminación: \n")
    sp.pprint(Ab)
    print("-" * 30)

    x3 = sp.simplify(Ab[2, 3] / Ab[2, 2])
    
    x2 = sp.simplify((Ab[1, 3] - Ab[1, 2]*x3) / Ab[1, 1])
    
    x1 = sp.simplify((Ab[0, 3] - Ab[0, 1]*x2 - Ab[0, 2]*x3) / Ab[0, 0])

    return x1, x2, x3

sol_x1, sol_x2, sol_x3 = eliminacion_gaussiana_simbolica()
print(f"x1 = {sol_x1}")
print(f"x2 = {sol_x2}")
print(f"x3 = {sol_x3}")
```

Para el inciso a, el valor de $\alpha$ que hace que el sistema no tenga soluciones es $\alpha = 1$ puesto que al sustituir este valor en la matriz triangular, se obtiene una fila que representa una contradicción (0 = 2), lo que indica que el sistema no tiene soluciones.

Para el inciso b, el valor de $\alpha$ que hace que el sistema tenga un número infinito de soluciones es $\alpha = -1$ puesto que al sustituir este valor en la matriz triangular, se obtiene una fila de ceros (0 = 0), lo que indica que el sistema tiene infinitas soluciones.

Para que exista una única solución como se propone en el inciso c, se puede elegir cualquier valor de $\alpha$ distinto de 1 y -1.

x = ($𝑥_j$) = [1000, 500, 350, 400], y 𝐛 = ($𝑏_𝑖$) = [3500, 2700, 900]. ¿Existe suficiente alimento para satisfacer el consumo promedio diario?

b. ¿Cuál es el número máximo de animales de cada especie que se podría agregar de forma individual al sistema
con el suministro de alimento que cumpla con el consumo?

```python
import numpy as np

A = np.array([
    [1, 2, 0, 3], 
    [1, 0, 2, 2],  
    [0, 0, 1, 1]   
], dtype=float)

x = np.array([1000, 500, 350, 400], dtype=float) 
b = np.array([3500, 2700, 900], dtype=float)    

consumo_actual = np.dot(A, x)
sobrante = b - consumo_actual

print(f"Suministro sobrante por recurso: {sobrante}")
print("-" * 40)

def calcular_incrementos_maximos(matriz_consumo, excedente_recursos):
    num_recursos, num_especies = matriz_consumo.shape
    resultados = []

    for j in range(num_especies):
        capacidades_por_recurso = []
        for i in range(num_recursos):
            consumo_unitario = matriz_consumo[i, j]
            
            if consumo_unitario > 0:
                capacidad = excedente_recursos[i] / consumo_unitario
                capacidades_por_recurso.append(capacidad)
        
        max_incremento = min(capacidades_por_recurso) if capacidades_por_recurso else float('inf')
        resultados.append(max_incremento)
    
    return resultados

incrementos = calcular_incrementos_maximos(A, sobrante)

for i, valor in enumerate(incrementos):
    print(f"Especie {i+1}: Puede incrementar un máximo de {valor:.0f} individuos.")
```

c. Si la especie 1 se extingue, ¿qué cantidad de incremento individual de las especies restantes se podría soportar?

```python
import numpy as np

def simular_extincion_y_crecimiento(especie_extinta_idx, A, x_actual, b_disponible):

    num_recursos, num_especies = A.shape
    poblacion_extinta = x_actual[especie_extinta_idx]
    alimento_recuperado = A[:, especie_extinta_idx] * poblacion_extinta
    
    x_restante = x_actual.copy()
    x_restante[especie_extinta_idx] = 0
    consumo_restante = np.dot(A, x_restante)
    
    nuevo_sobrante = b_disponible - consumo_restante
    
    print(f"--- Escenario: Extinción de Especie {especie_extinta_idx + 1} ---")
    print(f"Alimento recuperado por extinción: {alimento_recuperado}")
    print(f"Nuevo suministro libre total: {nuevo_sobrante}")
    
    incrementos = {}
    for j in range(num_especies):
        if j == especie_extinta_idx:
            continue
            
        capacidades = []
        for i in range(num_recursos):
            consumo_unitario = A[i, j]
            if consumo_unitario > 0:
                capacidad = nuevo_sobrante[i] / consumo_unitario
                capacidades.append(capacidad)
        
        max_inc = min(capacidades) if capacidades else float('inf')
        incrementos[f"Especie {j+1}"] = round(max_inc, 2)
        
    return incrementos

A = np.array([
    [1, 2, 0, 3],
    [1, 0, 2, 2],
    [0, 0, 1, 1]
], dtype=float)

x = np.array([1000, 500, 350, 400], dtype=float)
b = np.array([3500, 2700, 900], dtype=float)

res_c = simular_extincion_y_crecimiento(0, A, x, b)
print("Incrementos posibles:", res_c)
```

a. $$\frac{1}{4}x_1 + \frac{1}{5}x_2 + \frac{1}{6}x_3 = 9$$

$$\frac{1}{3}x_1 + \frac{1}{4}x_2 + \frac{1}{5}x_3 = 8$$

$$\frac{1}{2}x_1 + x_2 + 2x_3  = 8$$

```python
def gauss_jordan_32bit(A, b):
    A = A.astype(np.float32)
    b = b.astype(np.float32)
    n = len(b)
    
    Ab = np.hstack([A, b.reshape(-1, 1)]).astype(np.float32)
    
    print("Matriz aumentada inicial:\n", Ab)

    for i in range(n):
        max_row = np.argmax(np.abs(Ab[i:, i])) + i
        
        if Ab[max_row, i] == 0:
            raise ValueError("El sistema no tiene solución única.")

        if max_row != i:
            Ab[[i, max_row]] = Ab[[max_row, i]]
            print(f"Intercambio de fila {i+1} con fila {max_row+1}")

        pivot = Ab[i, i]
        Ab[i, :] = Ab[i, :] / pivot
        
        for j in range(n):
            if i != j: 
                factor = Ab[j, i]
                Ab[j, :] = Ab[j, :] - factor * Ab[i, :]
            
        print(f"Estado tras procesar columna {i+1}:\n", Ab)

    x = Ab[:, n]
    
    return x

A_a = np.array([[1/4, 1/5, 1/6], 
                [1/3, 1/4, 1/5], 
                [1/2, 1, 2]], dtype=np.float32)

b_a = np.array([9, 8, 8], dtype=np.float32)

print("--- RESOLVIENDO SISTEMA A MEDIANTE GAUSS-JORDAN ---")
sol_a = gauss_jordan_32bit(A_a, b_a)
print("\nSolución Sistema A:", sol_a)
```
